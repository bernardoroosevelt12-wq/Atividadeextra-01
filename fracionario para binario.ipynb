```python
#Jupyter: faça um programa que converta  um número fracionário decimal para binário, e vice-versa. Explicar em sala de aula.
def decimal_inteiro_para_binario(inteiro_decimal):
    if inteiro_decimal == 0:
        return "0"
    binario_str = ""
    while inteiro_decimal > 0:
        resto = inteiro_decimal % 2
        binario_str = str(resto) + binario_str
        inteiro_decimal = inteiro_decimal // 2
    return binario_str
def decimal_frac_para_binario(frac_decimal, precisao=10):
    if frac_decimal == 0:
        return "0"
    binario_frac = "" 
    for _ in range(precisao):
        frac_decimal *= 2
        bit = int(frac_decimal)
        binario_frac += str(bit)
        frac_decimal -= bit
        if frac_decimal == 0:
            break
    return binario_frac

valordecimal = float(input("insira um valor decimal "))
binario_inteiro = decimal_inteiro_para_binario(int(valordecimal))
binario_fracionario = decimal_frac_para_binario(abs(valordecimal - int(valordecimal)))
if binario_fracionario == "0":
    print (binario_inteiro)
else:
    print(f"{binario_inteiro}.{binario_fracionario}")
```
