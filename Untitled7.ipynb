```python
OCT_BIN_MAP = {
    '0': '000', '1': '001', '2': '010', '3': '011',
    '4': '100', '5': '101', '6': '110', '7': '111'
}
BIN_OCT_MAP = {
    '000': '0', '001': '1', '010': '2', '011': '3',
    '100': '4', '101': '5', '110': '6', '111': '7'
}
HEX_BIN_MAP = {
    '0': '0000', '1': '0001', '2': '0010', '3': '0011',
    '4': '0100', '5': '0101', '6': '0110', '7': '0111',
    '8': '1000', '9': '1001', 'A': '1010', 'B': '1011',
    'C': '1100', 'D': '1101', 'E': '1110', 'F': '1111'
}
BIN_HEX_MAP = {
    '0000': '0', '0001': '1', '0010': '2', '0011': '3',
    '0100': '4', '0101': '5', '0110': '6', '0111': '7',
    '1000': '8', '1001': '9', '1010': 'A', '1011': 'B',
    '1100': 'C', '1101': 'D', '1110': 'E', '1111': 'F'
}

def octal_para_binario(num_octal):
    """Converte um número octal (string) para binário."""
    if not isinstance(num_octal, str) or not all(c in '01234567' for c in num_octal):
        raise ValueError("Entrada inválida. O número octal deve ser uma string de dígitos de 0 a 7.")

    binario = "".join([OCT_BIN_MAP[digito] for digito in num_octal])
    return binario.lstrip('0') or '0' 

def hex_para_binario(num_hex):
    """Converte um número hexadecimal (string) para binário."""
    if not isinstance(num_hex, str) or not all(c in '0123456789ABCDEFabcdef' for c in num_hex):
        raise ValueError("Entrada inválida. O número hexadecimal deve ser uma string de dígitos válidos.")

    num_hex = num_hex.upper() 
    binario = "".join([HEX_BIN_MAP[digito] for digito in num_hex])
    return binario.lstrip('0') or '0' 

def binario_para_octal(num_binario):
    """Converte um número binário (string) para octal."""
    if not isinstance(num_binario, str) or not all(c in '01' for c in num_binario):
        raise ValueError("Entrada inválida. O número binário deve ser uma string de dígitos 0 ou 1.")

    padding_needed = (3 - len(num_binario) % 3) % 3
    num_binario_padded = '0' * padding_needed + num_binario

    octal = ""
    for i in range(0, len(num_binario_padded), 3):
        grupo_3_bits = num_binario_padded[i : i+3]
        octal += BIN_OCT_MAP[grupo_3_bits]

    return octal.lstrip('0') or '0' 

def binario_para_hex(num_binario):
    """Converte um número binário (string) para hexadecimal."""
    if not isinstance(num_binario, str) or not all(c in '01' for c in num_binario):
        raise ValueError("Entrada inválida. O número binário deve ser uma string de dígitos 0 ou 1.")

    padding_needed = (4 - len(num_binario) % 4) % 4
    num_binario_padded = '0' * padding_needed + num_binario

    hexadecimal = ""
    for i in range(0, len(num_binario_padded), 4):
        grupo_4_bits = num_binario_padded[i : i+4]
        hexadecimal += BIN_HEX_MAP[grupo_4_bits]

    return hexadecimal.lstrip('0') or '0' 

while True:
    print("\nEscolha uma opção de conversão:")
    print("1. Octal para Binario")
    print("2. Hexadecimal para Binário")


    
    print("3. Binário para Octal")
    print("4. Binário para Hexadecimal")
    

    escolha = input("Escolha sua conversao: ")

    if escolha == '1':
        num_entrada = input("Digite o número octal: ")
        try:
            resultado = octal_para_binario(num_entrada)
            decimal_result = int(resultado, 2)
            print(f"O octal {num_entrada} em binário é: {resultado}")
            print(f"O valor decimal correspondente é: {decimal_result}")
        except ValueError as e:
            print(f"Erro: {e}")
    elif escolha == '2':
        num_entrada = input("Digite o número hexadecimal: ")
        try:
            resultado = hex_para_binario(num_entrada)
            decimal_result = int(resultado, 2)
            print(f"O hexadecimal {num_entrada} em binário é: {resultado}")
            print(f"O valor decimal correspondente é: {decimal_result}")
        except ValueError as e:
            print(f"Erro: {e}")
    elif escolha == '3':
        num_entrada = input("Digite o número binário: ")
        try:
            resultado = binario_para_octal(num_entrada)
            print(f"O binário {num_entrada} em octal é: {resultado}")
        except ValueError as e:
            print(f"Erro: {e}")
    elif escolha == '4':
        num_entrada = input("Digite o número binário: ")
        try:
            resultado = binario_para_hex(num_entrada)
            print(f"O binário {num_entrada} em hexadecimal é: {resultado}")
        except ValueError as e:
            print(f"Erro: {e}")
```
