```python
#Faça um programa que o usuário entre com o número de níveis de quantização e receba como resposta o número de bits e todas as combinações desses bits.

import math
import itertools

quantizacao = int(input("insira o numero de quantizacao "))
if (math.log(quantizacao,2)).is_integer() == False:
 raise ValueError("precisa ser um numero na base 2 .coloque um numero na base 2.")
ndebits = math.log(quantizacao,2)
print ("o numero de bits e igual a "+str(ndebits))
for i in range(2 ** int(ndebits)):
  binario = format(i, f'{int(ndebits)}b')
  print(binario)
```
