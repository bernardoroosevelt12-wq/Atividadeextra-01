```python
#Jupyter: faça um programa que converta de binario para decimal. É para ser explicado em sala de aula.
binario_str = input("insira um valor binario: ")
decimal = 0
expoente = 0

for digit in reversed(binario_str):
    if digit == '1': 
        decimal += 2 ** expoente
    elif digit == '0':
        pass 
    else:
        raise ValueError(f"Entrada inválida: '{digit}' não é um dígito binário. Por favor, insira apenas 0s e 1s.")
      
    expoente += 1 

print (f"O valor decimal é: {decimal}")
```
